```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, Dropout
from scikeras.wrappers import KerasClassifier, KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
import keras as keras


previsores = pd.read_csv('DataBase/entradas_breast.csv')
classe = pd.read_csv('DataBase/saidas_breast.csv')
```

```python
def criarRede(optimizer, loss, kernel_initializer, activation, neurons):
    classificador = Sequential()
    classificador.add(Dense(units=neurons, activation=activation, kernel_initializer=kernel_initializer, input_dim=30))
    classificador.add(Dense(units=neurons, activation=activation, kernel_initializer=kernel_initializer))
    classificador.add(Dense(units=1, activation='sigmoid'))
    classificador.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
    return classificador

```

```python
from sklearn.model_selection import GridSearchCV 
from scikeras.wrappers import KerasClassifier

def criar_rede(optimizer, loss, kernel_initializer, activation, neurons):
    classificador = Sequential()
    classificador.add(Dense(units=neurons, activation=activation, kernel_initializer=kernel_initializer, input_dim=30))
    classificador.add(Dense(units=neurons, activation=activation, kernel_initializer=kernel_initializer))
    classificador.add(Dense(units=1, activation='sigmoid'))
    classificador.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
    return classificador

classificador = KerasClassifier(build_fn=criar_rede)

params={
    'batch_size':[20,25],
    'epochs':[10,20],
    'model__optimizer':['adam', 'sgd'],
    'model__loss': ['binary_crossentropy', 'mean_squared_error'],
    'model__kernel_initializer': ['random_uniform', 'normal'],
    'model__activation':['relu','softmax'],
    'model__neurons':[6,7],
    }



grid = GridSearchCV(estimator=classificador, param_grid=params)
grid_search = grid.fit(previsores, classe)



```

```python

melhores_parametros = grid.best_params_
melhor_precisao = grid.best_score_
print(melhores_parametros)
print(melhor_precisao)
```
